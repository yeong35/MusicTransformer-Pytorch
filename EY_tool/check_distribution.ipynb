```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mae = pd.read_csv("/home/bang/PycharmProjects/MusicGeneration/remi/maestro_split/output_0404.csv")
absolute = pd.read_csv("/home/bang/PycharmProjects/MusicGeneration/remi/MT_absolute/absolute_output_0404.csv")
classic = pd.read_csv("/home/bang/PycharmProjects/MusicGeneration/remi/MT_classic/classic_output_0404.csv")
```

```python
mae = mae[:-1]
mae
```

```python
absolute = absolute[:-1]
absolute
```

```python
classic = classic[:-1]
classic
```

```python
fig, axes = plt.subplots(2, 4, figsize=(30, 12))

for index, c_name in enumerate(classic.columns[1:]):
    axes[index//4][index%4].set_title(c_name, size=15)
    axes[index//4][index%4].hist([mae[c_name], classic[c_name], absolute[c_name]], label = ['mae', 'midi-like', 'relative'])
    axes[index//4][index%4].legend()
plt.show()
```

```python
for index, c_name in enumerate(classic.columns[1:]):
    print(f'{c_name} : MAESTRO = {mae[c_name].median()}, MIDI-like = {classic[c_name].median()}, relative = {absolute[c_name].median()}')
```
